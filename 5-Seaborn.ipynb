```python
import seaborn as sns
```

```python
df=sns.load_dataset("tips")
```

```python
df.shape
```

```python
df.head()
```

```python
df.dtypes
```

```python
#find the correlation with the dataset
df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

# kind='reg' will draw a regression line and probablity density function 

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

```python
# Pairlot/Scatter Plot
```

```python
sns.pairplot(df)
```

```python
#can use classification data too
sns.pairplot(df,hue='sex')
```

```python
df['sex'].value_counts()
```

# distplot for distribution of column feature

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

# Categorical Plot

```python
#countplot
```

```python
sns.countplot('smoker',data=df)
```

```python
sns.countplot('sex',data=df)
```

```python
sns.countplot(y='sex',data=df)
```

```python
#barplot, between two values, x and y. count plot gives count of a category. barplot compares between a numeric column and a categorical column. 
```

```python
sns.barplot(y='total_bill',x='sex',data=df)
```

```python
sns.barplot(x='day',y='total_bill',data=df)
```
