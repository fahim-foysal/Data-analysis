```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
#example
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
plt.scatter(x,y,c='g')
plt.xlabel('X axes')
plt.ylabel('Y axes')
plt.title('Graph in 2D')
plt.savefig('fig1.png')
```

```python
y=x*x
```

```python
plt.plot(x,y,'r*-')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*-')
plt.subplot(2,2,3)
plt.plot(x,y,'bo-')
plt.subplot(2,2,4)
plt.plot(x,y,'y*-')
```

```python
#bar plot
x=[2,8,10,12]
y=[5,7,12,13]
x2=[16,21,14,11]
y2=[18,19,20,22]
plt.bar(x,y)
plt.bar(x2,y2,color='g')
plt.title('Bar Graph')
plt.xlabel('X axes')
plt.ylabel('Y axes')
```

```python
#Box plot using matplotlib
```

```python
data=[np.random.normal(0,std,100) for std in range(0,4)]
```

```python
plt.boxplot(data,vert=True, patch_artist=True)
```

```python
#pie chart
labels='python','C++','java','C'
sizes=[215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.2, 0, 0,0.2)  # explode 1st slice

plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)
```
